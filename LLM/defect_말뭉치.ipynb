```python
import re
import numpy as np
import pandas as pd
import json
# 전처리를 위한 매핑 딕셔너리 정의
with open('/Users/admin/softeer/project/정리/LLM/label.json', 'r', encoding='utf-8') as file:
    mapping = json.load(file)
df = pd.read_csv(r'/Users/admin/softeer/project/정리/LLM/issue.csv')
df.head(2)
# 전처리 함수
def preprocess_data(data, mapping):
    preprocessed_data = []
    for item in data:
        if item is not np.nan:
            # 괄호 안의 값을 제거
            item_clean = re.sub(r'\([^)]*\)', '', item).strip()
            # " 기타" 제거
            item_clean = item_clean.replace(" 기타", "")
            # 매핑된 값 적용
            preprocessed_data.append(mapping.get(item_clean, item_clean))
        else:
            preprocessed_data.append(np.nan)
    return np.array(preprocessed_data)

preprocessed_data = preprocess_data(df['장치분류'], mapping)

# 전처리된 데이터에서 "nan" 문자열 제거
preprocessed_data = preprocessed_data[preprocessed_data != 'nan']

# 전처리된 데이터 확인
preprocessed_data_df = pd.DataFrame(preprocessed_data, columns=["전처리된 데이터"])
print(preprocessed_data_df['전처리된 데이터'].unique())
df[['장치분류']]=preprocessed_data_df[["전처리된 데이터"]]
df.to_csv('filtered_issue.csv')
df.head(2)
```
